## Verification des dataframes
            (verifier les duplications, et les valeurs null)

```python
def verification(df,name):
    d = df[df.duplicated()]
    Nan = df[df.isna().any(axis=1)]
    print('----------',name,' dataframe --------')
    print('Il y a {} lignes est repete'.format(len(d)))
    print('Il y a {} valeurs null'.format(len(Nan)))
    print()
    return d, Nan
```

# Data Preprocessing

## Numérisation
        (conversion des données symboliques vers numériques)

### 1. Determiner les caracteristiques symbolique
            (protocol_type, service, flag)

```python
Qualitative_col = ['protocol_type','service','flag']
df_train_symb = df_train[Qualitative_col]
df_test_symb = df_test[Qualitative_col]
```

```python
print('training data :')
df_train_symb.head(10)
```

```python
print('testing data :')
df_test_symb.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
def applyLabelEncoder(df):
    return df.apply(LabelEncoder().fit_transform)
```

```python
def dummyColonnesHeader(df):
    str = ['Protocol_','service_','flag_']
    protocol_type=sorted(df.protocol_type.unique())
    protocol_list=[str[0] + x for x in protocol_type]
    service=sorted(df.service.unique())
    service_list=[str[1] + x for x in service]
    flag=sorted(df.flag.unique())
    flag_list=[str[2] + x for x in flag]
    return protocol_list + service_list + flag_list


def applyOneHotEncoder(df_LbEn,df_symb):
    enc = OneHotEncoder()
    data_transform = enc.fit_transform(df_LbEn)
    return pd.DataFrame(data_transform.toarray(),columns=dummyColonnesHeader(df_symb))
```

### 2. Application LabelEncoder

```python
df_train_LbEn = applyLabelEncoder(df_train_symb)
df_test_LbEn = applyLabelEncoder(df_test_symb)
```

```python
print('Training dataframe apres application de LabelEncoder :')
df_train_LbEn.head(10)
```

```python
print('testing dataframe apres application de LabelEncoder :')
df_test_LbEn.head(10)
```

## traitement à propos colonnes des attacks

```python
def Attacks(attack):
    Dos = ['neptune','land','pod','smurf','teardrop','back','worm','udpstorm','processtable','apache2']
    Probe = ['ipsweep','satan','nmap','portsweep','mscan','saint']
    R2l = ['ftp_write','guess_passwd','imap','multihop','phf','spy','warezclient','warezmaster','snmpguess','named','xlock','xsnoop','snmpgetattack','httptunnel','sendmail']
    U2r = ['buffer_overflow','loadmodule','perl','rootkit','ps','xterm','sqlattack']
    if attack in Dos:
        atk = 1
    elif attack in Probe:
        atk = 2
    elif attack in R2l:
        atk = 3
    elif attack in U2r:
        atk = 4
    else:
        atk = 0
    return atk
```

```python
train_attack_types = df_train_new.attack.apply(Attacks)
test_attack_types = df_test_new.attack.apply(Attacks)
```

```python
df_train_new['attack'] = train_attack_types
df_test_new['attack'] = test_attack_types
```

```python
print("Training data : ")
print(df_train_new.attack.value_counts())
print("\ntesting data : ")
print(df_test_new.attack.value_counts())
```

```python
def DiagAttack(df, name):
    ax = df.attack.value_counts()
    ax.plot(kind='pie',labels=None,subplots=True,figsize=(5,8))
    plt.legend(loc='center left',labels=['Normal','DOS','Probe','R2L','U2R'],bbox_to_anchor=(1, 0.5))
    plt.figtext(.5,.8,'Attack pour '+str(name)+' dataframe',fontsize=10, ha='center')
```

```python
DiagAttack(df_train_new,'training')
```

```python
DiagAttack(df_test_new,'testing')
```
